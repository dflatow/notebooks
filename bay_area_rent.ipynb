```python
import requests
import bs4
import re
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pylab as plt
pd.options.display.float_format = '{:20,.0f}'.format

def flatten(x):
    return [item for sublist in x for item in sublist]

def process_data(d):
    d['beds_int'] = d['beds'].astype(int)
    d['rent'] = d['rent'].astype(int)
    d['p'] = d['rent'] / d['beds_int']
    d = d[(d['p'] > 0) & (d['p'] < 5000)]
    return d

parser = re.compile("\$(?P<rent>\d+)\s+/\s+(?P<beds>\d)")
def parse_raw(r):
    return pd.Series(parser.search(r).groupdict())

def get_data(root):
    pgs = ['']
    lim = 2000
    pgs.extend(["&s=" + str(x) for x in range(100, lim, 100)])
    responses = [requests.get(root + pg) for pg in pgs]
    selections = flatten([bs4.BeautifulSoup(r.text).select("#searchform > div.rightpane > div.content > p > span > span.l2") 
                for r in responses])
    data = [s.text for s in selections]
    assert(len(data) < lim)
    print "got %d records" % len(data)
    return pd.concat([parse_raw(r) for r in data], axis= 1).T

def summary(data):
    
    data[['p', 'beds']].groupby("beds").boxplot('p')
    plt.scatter(data.beds_int +  np.random.randn(len(data)) / 20 , data.p)
    
    fns = ['mean', 'median', 'count']
    sum_stats = pd.concat([pd.pivot_table(data, 'p', 'beds', aggfunc=f) for f in fns], axis=1)
    sum_stats.columns = fns
     
    return data.groupby("beds")['p'].describe().unstack()
```
